# A Comparison of Log Scale Chart to a Linear Scale Chart

## Why should you prefer a Log Scale Chart?
A Linear Scale Chart does not provide an accurate visual of the price change value of an asset. It amplifies price changes on the chart where the absolute prices are of a greater magnitude. For eg. you will see a much bigger run up on the chart when an asset moves from 1000 to 1500, than when it moves from 10 to 15; although the returns in both cases are the same, i.e. 50%.

```python
# Imprting the requried libraries
import pandas as pd
import yfinance as yf
from matplotlib import pyplot as plt
import datetime as dt
```

```python
# Let us download the BSE Sensex Index Price Data throughout its available history
now = dt.datetime.now()
bse = yf.download('^BSESN', start = '1997-01-01', end = now, progress = False)
```

```python
# Renaming the Adj Close column to get rid of the space within the column heading
bse.rename(columns={'Adj Close':'Adj_Close'}, inplace=True)
```

```python
# Let us list the oldest 5 recaords to see the earliest dates
bse.head()
```

```python
# Let us list the latest 5 recaords to see if we have the latest data
bse.tail()
```

#### Let us plot the same data (Adjusted Closing Price of BSE across years) on a linear scale as well as a log scale.

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 8), layout='constrained')  # Defining the figure and its size.
ax[0].plot(bse['Adj_Close'], label='linear')  # Plot Adjusted Close on a line graph.
ax[0].set_xlabel('Years')  # Add an x-label to the axes.
ax[0].set_ylabel('BSE Index')  # Add a y-label to the axes.
ax[0].set_title("BSE Sensex - Linear Scale")  # Add a title to the axes.

ax[1].set_yscale('log', base = 10)  # Set the y-axis scale to log with Base 10 instead of Linear.
ax[1].plot(bse['Adj_Close'], label='linear')  # Plot Adjusted Close on a line graph.
ax[1].set_xlabel('Years')  # Add an x-label to the axes.
ax[1].set_ylabel('BSE Index')  # Add a y-label to the axes.
ax[1].set_title("BSE Sensex - Log Scale")  # Add a title to the axes.
```

#### To illustrate my point further, let us focus on 2 aspects of the data.
1. The bull run from 2003 to 2007
2. The bull run from 2020 to 2021

```python
# Lets download the data for these 2 periods
bse_0307 = yf.download('^BSESN', start = '2003-07-01', end = '2007-12-01', progress = False)
bse_2021 = yf.download('^BSESN', start = '2020-04-01', end = '2021-12-31', progress = False)
```

#### If we inspect BSE Sensex price increase from April 2003 to December 2007, we see that the BSE Sensex gained a 61.3% CAGR going from 2094 to 20287.

#### Now if we inspect the price increase for BSE Sensex from April 2020 to September 2021, BSE Sensex gained a CAGR of 50% going from 27590 to 49035.

#### Although the growth is the same, visually, the 2020-21 rise looks dramatic and steep compared to the 2003-07 rise on the linear scale graph.

#### Let us look at the bull run from 2003 to 2007 in linear and log scaled graphs.

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 8), layout='constrained')  # Defining the figure and its size.
ax[0].plot(bse_0307['Adj Close'], label='linear')  # Plot Adjusted Close on a line graph.
ax[0].set_xlabel('Years')  # Add an x-label to the axes.
ax[0].set_ylabel('BSE Index')  # Add a y-label to the axes.
ax[0].set_title("BSE Sensex (2003 to 2007) - Linear Scale")  # Add a title to the axes.

ax[1].set_yscale('log', base = 10)  # Set the y-axis scale to log with Base 10 instead of Linear.
ax[1].plot(bse_0307['Adj Close'], label='linear')  # Plot Adjusted Close on a line graph.
ax[1].set_xlabel('Years')  # Add an x-label to the axes.
ax[1].set_ylabel('BSE Index')  # Add a y-label to the axes.
ax[1].set_title("BSE Sensex (2003 to 2007) - Log Scale")  # Add a title to the axes.
```

And this is BSE Sensex bull run from 2020 to 2021 in Linear and Log Scales

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 8), layout='constrained')  # Defining the figure and its size.
ax[0].plot(bse_2021['Adj Close'], label='linear')  # Plot Adjusted Close on a line graph.
ax[0].set_xlabel('Years')  # Add an x-label to the axes.
ax[0].set_ylabel('BSE Index')  # Add a y-label to the axes.
ax[0].set_title("BSE Sensex (2020 to 2021) - Linear Scale")  # Add a title to the axes.

ax[1].set_yscale('log', base = 10)  # Set the y-axis scale to log with Base 10 instead of Linear.
ax[1].plot(bse_2021['Adj Close'], label='linear')  # Plot Adjusted Close on a line graph.
ax[1].set_xlabel('Years')  # Add an x-label to the axes.
ax[1].set_ylabel('BSE Index')  # Add a y-label to the axes.
ax[1].set_title("BSE Sensex (2020 to 2021) - Log Scale")  # Add a title to the axes.
```

#### The above two graph sets show a visually similar graph between linear and log scales. This is due to the fact that there will not be much visual difference when the timeframe is short. The actual difference in scaling is revealed at a much longer timeframe.
